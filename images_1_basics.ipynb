# Load + show image

```python
def show(im):
    return imshow(np.asarray(im))

path_in = 'data/input/heart_prism.jpg'
im = Image.open(path_in)
show(im)
```

# Select box

```python
box = (0, 0, 48, 48)
im_box = im.crop(box)
im_box
```

# Copy box

```python
box = (0, 0, 100, 100)
im_box = im.crop(box).copy()
im_box
```

# Resize 

```python
im.crop((0,0,20,20)).resize((100,100))
```

# Get pixel

```python
im.getpixel((10,10))
```

# Put pixel

```python
box = (0, 0, 60, 60)
im_box = im.crop(box).copy()
for x in range(0,im_box.size[0]):
    for y in range(0,im_box.size[1]):
        im_box.putpixel((x,y),(0,0,0))
im_box
```

# Set color

```python
def set_color(im,color):
    for x in range(im.size[0]):
        for y in range(im.size[1]):
            im.putpixel((x,y),color)
    return im
im_small = im.crop((0,0,100,100)).copy()
set_color(im_small,(0,0,0))
```

# Paste

```python
# Copy the left half and paste over right half
x_mid = int(im.size[0]/2)
y1 = int(im.size[1])
im_half = im.copy()
im_half.paste(im_half.crop((0,0,x_mid,y1)),(x_mid,0,x_mid*2,y1))
show(im_half)
```
